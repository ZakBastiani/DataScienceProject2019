```python
from sklearn.externals import joblib 
import pandas as pd
import numpy as np
loaded_model = joblib.load('MovieSuccessRate.pkl')
data = pd.DataFrame(pd.read_csv("movies4app.csv",  encoding="utf-8"))
```

```python
genres = data["Genres"].unique()
cast1data = data["Cast1"].unique()
cast2data = data["Cast2"].unique()
cast3data = data["Cast3"].unique()
cast4data = data["Cast4"].unique()
dir1data = np.array(data["Director_One"].unique())
dir2data = data["Director_two"].unique()
dir3data = data["Director_three"].unique()
cast1data = cast1data.astype(str)
cast2data = cast2data.astype(str)
cast3data = cast3data.astype(str)
cast4data = cast4data.astype(str)
dir1data = dir1data.astype(str)
dir2data = dir2data.astype(str)
dir3data = dir3data.astype(str)
cast1data = np.sort(cast1data)
cast2data = np.sort(cast2data)
cast3data = np.sort(cast3data)
cast4data = np.sort(cast4data)
dir1data = np.sort(dir1data)
dir2data = np.sort(dir2data)
dir3data = np.sort(dir3data)
```

```python
from tkinter import *
 
from tkinter.ttk import *
 
window = Tk()
 
window.title("Movie Testing App")
 
window.geometry('800x300')
     
ratinglbl = Label(window, text="Rating: ", font=("Arial", 15))
ratinglbl.grid(column=0, row=0)

rating = Combobox(window)
rating['values']= ('G', 'PG', "PG13", "R", "Unrated")
rating.current(1) 
rating.grid(column=1, row=0)

genrelbl = Label(window, text="Genre: ", font=("Arial", 15))
genrelbl.grid(column=0, row=1)

genre = Combobox(window)
genre['values']= genres.tolist()
genre.current(1)
genre.grid(column=1, row=1)

castlbl = Label(window, text="Cast: ", font=("Arial", 15))
castlbl.grid(column=0, row=2)

cast1 = Combobox(window)
cast1['values']= cast1data.tolist()
cast1.current(1) 
cast1.grid(column=1, row=2)

cast2 = Combobox(window)
cast2['values']= cast2data.tolist()
cast2.current(1) 
cast2.grid(column=2, row=2)

cast3 = Combobox(window)
cast3['values']= cast3data.tolist()
cast3.current(1) 
cast3.grid(column=3, row=2)

cast4 = Combobox(window)
cast4['values']= cast4data.tolist()
cast4.current(1) 
cast4.grid(column=4, row=2)

dirlbl = Label(window, text="Directors: ", font=("Arial", 15))
dirlbl.grid(column=0, row=3)

dir1 = Combobox(window)
dir1['values']= dir1data.tolist()
dir1.current(0) 
dir1.grid(column=1, row=3)

dir2 = Combobox(window)
dir2['values']= dir2data.tolist()
dir2.current(0) 
dir2.grid(column=2, row=3)

dir3 = Combobox(window)
dir3['values']= dir3data.tolist()
dir3.current(0) 
dir3.grid(column=3, row=3)

results = Label(window, text="Result: ", font=("Arial", 15))
results.grid(column=1, row=4)

output = Label(window, text=" ", font=("Arial", 15))
output.grid(column=2, row=4) 

def test():
    #Build the array of the info
    inputdata = [rating.get(), genre.get(),cast1.get(), cast2.get(), cast3.get(), cast4.get(), dir1.get(), dir2.get(), dir3.get()]
    test = pd.DataFrame([inputdata] ,columns= ['Rating', 'Genres', 'Cast1', 'Cast2' ,'Cast3', 'Cast4','Director_One','Director_two','Director_three'])
    #
    word = test["Rating"][0]
    List = data[data["Rating"] == word]
    test["Rating"][0] = List['Profit'].mean()
    word = test["Genres"][0]
    genList = data[data["Genres"] == word]
    test["Genres"][0] = List['Profit'].mean()
    word = test["Cast1"][0]
    List = data[data["Cast1"] == word]
    test["Cast1"][0] = List['Profit'].mean()
    word = test["Cast2"][0]
    List = data[data["Cast2"] == word]
    test["Cast2"][0] = List['Profit'].mean()
    word = test["Cast3"][0]
    List = data[data["Cast3"] == word]
    test["Cast3"][0] = List['Profit'].mean()
    word = test["Cast4"][0]
    List = data[data["Cast4"] == word]
    test["Cast4"][0] = List['Profit'].mean()
    word = test["Director_One"][0]
    List = data[data["Director_One"] == word]
    test["Director_One"][0] = List['Profit'].mean()
    word = test["Director_two"][0]
    List = data[data["Director_two"] == word]
    test["Director_two"][0] = List['Profit'].mean()
    word = test["Director_three"][0]
    List = data[data["Director_three"] == word]
    test["Director_three"][0] = List['Profit'].mean()
    
    success = loaded_model.predict(test)
    print(success)
    if(success[0]):
        output.configure(text= "Success")
    else:
        output.configure(text= "Failure")

btn = Button(window, text="Test", command=test)
btn.grid(column=0, row=4)


window.mainloop()
```
